```python
import matplotlib.pyplot as plt
xs=[0]*129
xs[65]=1
size=129
xst=[xs]
ts=64
def transtition(l,c,f):
    match(l,c,f):
     case(0,0,0):return 0
     case(0,0,1):return 1
     case(0,1,0):return 1
     case(0,1,1):return 1
     case(1,0,0):return 1
     case(1,0,1):return 0
     case(1,1,0):return 0
     case(1,1,1):return 0
for t in range(ts):
   xs_next=[0]*129
   for x in range(1,size-1):
      xs_next[x]=transtition(xs[x-1],xs[x],xs[x+1])
   xs=xs_next
   xst.append(xs_next)
fig,ax=plt.subplots()
ax.invert_yaxis()
plt.pcolor(xst,cmap='binary')
plt.show()
```
