```python
import matplotlib.pyplot as plt
for n in range(256):
    xs=[0]*30
    xstz=[]
    xs[15]=1
    size=30
    xst=[xs]
    ts=50
    def transtition(l,c,f):
        match(l,c,f):
         case(0,0,0):return n%2
         case(0,0,1):return (n%4)//2
         case(0,1,0):return (n%8)//4
         case(0,1,1):return (n%16)//8
         case(1,0,0):return (n%32)//16
         case(1,0,1):return (n%64)//32
         case(1,1,0):return (n%128)//64
         case(1,1,1):return n//128
    for t in range(ts):
        xs_next=[0]*30
        for x in range(1,size-1):
         xs_next[x]=transtition(xs[x-1],xs[x],xs[x+1])
        xs=xs_next
        xst.append(xs_next)
    fig,ax=plt.subplots()
    ax.invert_yaxis()
    plt.pcolor(xst,cmap='binary')
    plt.show()
```
