```python
import matplotlib.pyplot as plt
import math
fig=plt.figure(figsize=(16,10))
def random_valueliner(x,a,b,M):             
    return a*x+b-((a*x+b)//M)*M
C=1
repeat=300
x1=[]
x1.append(C)
for i in range(repeat):
    C=random_valueliner(C,1103515245,12345,2**32)       #Unix
    x1.append(C)
plt.subplot(2,2,1)
plt.plot(x1)
plt.subplot(2,2,2)
plt.plot([x1[i+1]-x1[i] for i in range(298)])
```
