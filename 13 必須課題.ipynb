```python
import matplotlib.pyplot as plt
import math
def random_valueliner(x,a,b,M):
    return a*x+b-((a*x+b)//M)*M
C=1
repeat=300
x1=[]
x1.append(C)
for i in range(repeat):
    C=random_valueliner(C,75,74,2**16+1)
    x1.append(C)
plt.subplot(2,2,1)
plt.plot(x1)
plt.subplot(2,2,2)
plt.plot([x1[i+1]-x1[i] for i in range(298)])
```

```python
k=5
x=[]
r=0
sum=0
vals=[]
for i in range(repeat):
    r=x1[i]/2**16
    sum+=r**k
    x.append(i)
    vals.append(math.sqrt(i)*abs(sum/(i+1)-1/(k+1)))
plt.plot(x,vals,color='green',marker='o')
plt.xlabel('index',fontsize=20)
plt.ylabel('value',fontsize=20)
plt.tick_params(labelsize=15)
plt.show()
```
