```python
import math
import matplotlib.pyplot as plt
x0=1
vx0=0
n=100
t=20
e=0.00001
ts1=[]
xs1=[]
def derivative(x,vx):
    dxdt=vx
    dvxdt=-2*x
    return dxdt,dvxdt
x=x0
vx=vx0
delta=(t)/n
for i in range(n):
    kx,kvx=derivative(x,vx)
    vx=vx+delta*kvx
    x=x+delta*kx
    xs1.append(x)
    t=i*delta+delta
    ts1.append(t)


xs2=[]
x=x0
vx=vx0
for i in range(n):
    k1x,k1vx=derivative(x,vx)
    k2x,k2vx=derivative(x+delta*k1x/2,vx+delta*k1vx/2)
    x=x+k2x*delta
    vx=vx+k2vx*delta
    xs2.append(x)
xs3=[]
x=x0
vx=vx0
for i in range(n):
    k1x,k1vx=derivative(x,vx)
    k2x,k2vx=derivative(x+k1x*delta/2,vx+k1vx*delta/2)
    k3x,k3vx=derivative(x+2*delta*k2x-delta*k1x,vx+2*delta*k2vx-delta*k1vx)
    x=x+delta*(k1x+4*k2x+k3x)/6
    vx=vx+delta*(k1vx+4*k2vx+k3vx)/6
    t=i*delta+delta
    xs3.append(x)
xs4=[]
x=x0
vx=vx0
for i in range(n):
    k1x,k1vx=derivative(x,vx)
    k2x,k2vx=derivative(x+k1x*delta/2,vx+k1vx*delta/2)
    k3x,k3vx=derivative(x+delta*k2x/2,vx+delta*k2vx/2)
    k4x,k4vx=derivative(x+delta*k3x,vx+delta*k3vx)
    x=x+delta*(k1x+2*k2x+2*k3x+k4x)/6
    vx=vx+delta*(k1vx+2*k2vx+2*k3vx+k4vx)/6
    xs4.append(x)
plt.plot(ts1,xs1,color='red',marker='o',label='Euler')
plt.plot(ts1,xs2,color='green',label='2nd-order')
plt.plot(ts1,xs3,color='yellow',label='3rd-order')
plt.plot(ts1,xs4,color='blue',label='4th-order')
plt.xlabel('t',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.legend(fontsize=15)
plt.tick_params(labelsize=15)
plt.show()
ex1=[]
ex2=[]
ex3=[]
ex4=[]
for i in range(n):
    a=abs((xs1[i]-math.cos(math.sqrt(2)*delta*i))/(math.cos(math.sqrt(2)*delta*i)+e))
    b=abs((xs2[i]-math.cos(math.sqrt(2)*delta*i))/(math.cos(math.sqrt(2)*delta*i)+e))
    c=abs((xs3[i]-math.cos(math.sqrt(2)*delta*i))/(math.cos(math.sqrt(2)*delta*i)+e))
    d=abs((xs4[i]-math.cos(math.sqrt(2)*delta*i))/(math.cos(math.sqrt(2)*delta*i)+e))
    ex1.append(a)
    ex2.append(b)
    ex3.append(c)
    ex4.append(d)
plt.plot(ts1,ex1,color='red',label='Euler')
plt.plot(ts1,ex2,color='green',label='2nd-order')
plt.plot(ts1,ex3,color='yellow',label='3rd-order')
plt.plot(ts1,ex4,color='blue',label='4th-order')
plt.yscale('log')
plt.xlabel('t',fontsize=20)
plt.ylabel('relative error',fontsize=20)
plt.legend(fontsize=15)
plt.tick_params(labelsize=15)
plt.show()
```
