```python
import math 
import matplotlib.pyplot as plt
a=0
n=0
x=0
def asin(n,x):
    return -(2/n)*math.sin(n*x)
def n1(n,x):
    return n**-2*math.sin(n*x)
def n2(n,x):
    return n**-2*x


def fourier(x,a,f):
    b=0.0
    for i in range(int(a)):
        b=b+f(i+1,x)
        if i==a-1:
         b=b+math.pi
    return b
list_x=list(math.pi/40*n for n in range(81) )
list_y=list(map(lambda x:fourier(x,10,asin),list_x))
def list_x1(l,n):
    return list(l/n*x for x in range(n))
def list_y1(l,f,n):
    return list(map(lambda x:fourier(x,10,f),list_x1(l,n)))
plt.plot(list_x1(1,50),list_y1(1,n2,50))

plt.plot(list_x,list_y)
```
