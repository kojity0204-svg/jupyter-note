```python
import math
import matplotlib.pyplot as plt
from matplotlib import animation,rc
from IPython.display import HTML
rc('animation', html='jshtml')
fig,ax=plt.subplots()
c=1.0
def Lorents_x(x,t,v):
    return x/math.sqrt(1-(v/c)**2)+c*t*(v/c)/math.sqrt(1-(v/c)**2)
def Lorents_t(x,t,v):
    return c*t/math.sqrt(1-(v/c)**2)-x*(v/c)/math.sqrt(1-(v/c)**2)
t=0
n=100
delta=0.1
a=delta/n
tra_x=[]
tra_y=[]
delta_t=10
n_t=10 
k=delta_t/n_t
def drawing():
    image=plt.plot(tra_x,tra_y)
    return image
history=[]
for t in range(n_t):
   tra_x=[]
   tra_y=[] 
   for i in range(n):
      b=int(n*math.sqrt(1-(i/n)**2))
      for s in range(b):                    #o~10で考える
         tra_x.append(Lorents_x(i*a,k*t,c/1.1))
         tra_x.append(Lorents_x(-i*a,k*t,c/1.1))
         tra_x.append(Lorents_x(-i*a,k*t,c/1.1))
         tra_x.append(Lorents_x(i*a,k*t,c/1.1))
         tra_y.append(s*a)
         tra_y.append(-s*a)
         tra_y.append(s*a)
         tra_y.append(-s*a)
   plt.plot(tra_x,tra_y)
   history.append(drawing())
ani=animation.ArtistAnimation(fig,history,interval=500)
plt.show()
ani
```
