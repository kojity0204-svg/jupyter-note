```python
import math
import matplotlib.pyplot as plt
def dec(x):
    return math.exp(-0.05*x)*math.sin(x)
def dec_differ(x):
    return 25/(100**2)*math.exp(-0.05*x)*math.sin(x)-0.1*math.exp(-0.05*x)*math.cos(x)-math.exp(-0.05*x)*math.sin(x)
def trip(x,f,h):
    if h==0:
        return 0
    else:
        return (f(x+h)-2*f(x)+f(x-h))/h**2    
def list_x(x,X,n):
    return list(x+X/n*i for i in range(n))

plt.subplot(2,2,1)
list_y1=list(map(lambda x:dec_differ(x),list_x(1,50,1000)))
list_y2=list(map(lambda x:trip(x,dec,0.01),list_x(1,50,1000)))
plt.plot(list_x(1,50,1000),list_y1,color='green')
plt.plot(list_x(1,50,1000),list_y2,color='yellow')
list_y22=list(map(lambda x: abs(dec_differ(x)-trip(x,dec,0.01))/dec_differ(x),list_x(1,50,1000)))
plt.subplot(2,2,2)
plt.plot(list_x(1,50,1000),list_y22,color='black')
plt.yscale('log')
```
