```python
import math
import matplotlib.pyplot as plt
def dec(x):
    return math.exp(-0.05*x)*math.sin(x)
def dec_differ(x):
    return math.exp(-0.05*x)*math.cos(x)-0.05*math.exp(-0.05*x)*math.sin(x)

def ad_differ(x,dx,f):
    if dx==0:
        return 0
    return (f(x+dx)-f(x))/dx
def cent_differ(x,dx,f):
    if dx==0:
        return 0
    return (f(x+dx)-f(x-dx))/(2*dx)
def list_x(x,X,n):
    return list(x+X/n*i for i in range(n))
plt.subplot(2,2,1)
list_y1=list(map(lambda x:dec_differ(x),list_x(0,2*math.pi,1000)))
list_y2=list(map(lambda x:ad_differ(x,0.01,dec),list_x(0,2*math.pi,1000)))
list_y3=list(map(lambda x:cent_differ(x,0.005,dec),list_x(0,2*math.pi,1000)))
plt.plot(list_x(0,2*math.pi,1000),list_y1,color='green')
plt.plot(list_x(0,2*math.pi,1000),list_y2,color='yellow')
plt.plot(list_x(0,2*math.pi,1000),list_y3,color='blue')
plt.subplot(2,2,2)
list_y22=list(map(lambda x: abs(dec_differ(x)-ad_differ(x,0.01,dec))/dec_differ(x),list_x(0,2*math.pi,1000)))
list_y32=list(map(lambda x: abs(dec_differ(x)-cent_differ(x,0.005,dec))/dec_differ(x),list_x(0,2*math.pi,1000)))
plt.plot(list_x(0,2*math.pi,1000),list_y22,color='black')
plt.plot(list_x(0,2*math.pi,1000),list_y32,color='purple')
plt.yscale('log')
```
