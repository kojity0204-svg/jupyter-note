```python
import math
import matplotlib.pyplot as plt
e=00000000000.1
a=0
b=0
repeat=5000
def Euler(f,x1,x2,t0,tend):             #x1x2
    v1=x2
    x01=x1
    delta=(tend-t0)/repeat
    for i in range(repeat):
        x01=x01+v1*delta
        v1=v1+f(x01)*delta
    return x01
def t_scale(t0,tend):
    deltat=(tend-t0)/repeat
    return  [ i*deltat for i in range(repeat)]
def f1(x):
    return -2*x
listf11=list(map(lambda t: Euler(f1,1,0,0,t),t_scale(0,50)))
plt.subplot(2,2,1)
plt.plot(t_scale(0,50),listf11)
def Ecent(f,x1,x2,t0,tend):             #x1x2
    v1=x2
    x01=x1
    deltat=(tend-t0)/repeat
    for i in range(repeat):
        a=x01
        x01=x01+v1*deltat
        b=x01
        v1=v1+f(x01+(b-a)/2)*deltat
    return x01
listf12=list(map(lambda t: Ecent(f1,1,0,0,t),t_scale(0,50)))
plt.plot(t_scale(0,50),listf12)
def analy(x):
    return math.sin(math.sqrt(2)*x)
def erroreu(x1,x2,t0,t):
    return (analy(t)-Euler(f1,x1,x2,t0,t))/(analy(t)+e)
def errorce(x1,x2,t0,t):
    return (analy(t)-Ecent(f1,x1,x2,t0,t))/(analy(t)+e)
plt.subplot(2,2,2)
listf13=list(map(lambda t: erroreu(1,0,math.pi/(math.sqrt(2)*2),t),t_scale(math.pi/(math.sqrt(2)*2),math.pi/(math.sqrt(2)))))
plt.plot(t_scale(0,50),listf13)
listf14=list(map(lambda t: errorce(1,0,math.pi/(math.sqrt(2)*2),t),t_scale(math.pi/(math.sqrt(2)*2),math.pi/(math.sqrt(2)))))
plt.plot(t_scale(0,50),listf14)
```
