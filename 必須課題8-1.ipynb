```python
import math 
import matplotlib.pyplot as plt
import numpy as np
delta=10/100
n0=100
datasetx=[1,2,3,4,5,6,7,8,9]
datasety=[10,15,18,27,40,32,41,44,60]
n=len(datasetx)
sumx=sum(datasetx)
sumx2=sum(xi**2 for xi in datasetx )
sumy=sum(datasety)
sum2y=sum(yi**2 for yi in datasety )
sumxy=sum(xi*yi for xi,yi in zip(datasetx,datasety))
a1=(n*sumxy-sumx*sumy)/(n*sumx2-sumx**2)
a2=sumy/n-a1*sumx/n
line=[a1*delta*i+a2 for i in range(n0)]
xaxis=[delta*i for i in range(n0)]
plt.plot(xaxis,line,color='green')
plt.scatter(datasetx,datasety,color='red')
```

一つ目の課題の一次近似によるフィッテイング
授業のコードを参考にした

```python
sumx3=sum(xi**3 for xi in datasetx )
sumx4=sum(xi**4 for xi in datasetx)
sumx2y=sum(xi**2*yi for xi,yi in zip(datasetx,datasety))
nd_x=np.array(datasetx)
nd_y=np.array(datasety)
nd1=np.array([[sumx2y],[sumxy],[sumy]])
nd_A=np.array([\
 [sumx4, sumx3, sumx2],\
 [sumx3, sumx2, sumx],\
 [sumx2, sumx, len(datasetx)]\
    ])
nd_Ainv=np.linalg.inv(nd_A)#逆行列
a=nd_Ainv@nd1#行列積
a=np.linalg.solve(nd_A, nd1)#上記2行と同じ
a1=a[0]
a2=a[1]
a3=a[2]
nd_ys_fit=[a1*xi**2+a2*xi+a3 for xi in datasetx]
plt.scatter(nd_x,nd_y,label='data',color='green')#numpyベクトルを指定可能
plt.plot(nd_x,nd_ys_fit,label='fit',color='magenta')#numpyベクトルを指定可能
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.legend(fontsize=15)
plt.tick_params(labelsize=15)
plt.show()
```

```python
datasety=[10,15,18,27,40,32,41,44,60]
logy=[math.log(yi) for yi in datasety ]
sumx=sum(datasetx)
sumx2=sum(xi**2 for xi in datasetx )
sumy=sum(logy)
sum2y=sum(yi**2 for yi in logy )
sumxy=sum(xi*yi for xi,yi in zip(datasetx,logy))
a1=(n*sumxy-sumx*sumy)/(n*sumx2-sumx**2)
a2=sumy/n-a1*sumx/n
line=[a1*delta*i+a2 for i in range(n0)]
xaxis=[delta*i for i in range(n0)]
plt.plot(xaxis,line,color='green')
plt.scatter(datasetx,logy,color='red')
```

一つ目の課題のlogによる近似を行った

```python
import math
import matplotlib.pyplot as plt
xs=[10,20,30,40,50,60]
ys=[20*math.exp(-0.2*xi)+0.01*math.sin(xi)+0.01*(math.cos(xi))**2+0.02 for xi in xs]
delta=60/100
n0=100
n=len(xs)
sumx=sum(xi for xi in xs)
sumx2=sum(xi**2 for xi in xs)
sumxy=sum(xi*yi for xi,yi in zip(xs,ys))
sumy=sum(ys)
sum2y=sum(yi**2 for yi in ys )
a1=(n*sumxy-sumx*sumy)/(n*sumx2-sumx**2)
a2=sumy/n-a1*sumx/n
line=[a1*delta*i+a2 for i in range(n0)]
xaxis=[delta*i for i in range(n0)]
plt.plot(xaxis,line,color='green')
plt.scatter(xs,ys,color='red')
```

二つ目の課題の一次近似を行った
関数にexp,sin,cosが入っているため一時での近似はさすがに無理があった

```python
import numpy as np
sumx3=sum(xi**3 for xi in xs )
sumx4=sum(xi**4 for xi in xs)
sumx2y=sum(xi**2*yi for xi,yi in zip(xs,ys))
nd_x=np.array(xs)
nd_y=np.array(ys)
nd1=np.array([[sumx2y],[sumxy],[sumy]])
nd_A=np.array([\
 [sumx4, sumx3, sumx2],\
 [sumx3, sumx2, sumx],\
 [sumx2, sumx, len(xs)]\
    ])
nd_Ainv=np.linalg.inv(nd_A)#逆行列
a=nd_Ainv@nd1#行列積
a=np.linalg.solve(nd_A, nd1)#上記2行と同じ
a1=a[0]
a2=a[1]
a3=a[2]
nd_ys_fit=[a1*xi**2+a2*xi+a3 for xi in xs]
plt.scatter(nd_x,nd_y,label='data',color='green')#numpyベクトルを指定可能
plt.plot(nd_x,nd_ys_fit,label='fit',color='magenta')#numpyベクトルを指定可能
plt.xlabel('x',fontsize=20)
plt.ylabel('y',fontsize=20)
plt.legend(fontsize=15)
plt.tick_params(labelsize=15)
plt.show()
```

必須課題二つ目の二次近似を行った
授業で扱われた手法を参考にした

```python
logy=[math.log(yi) for yi in ys ]
sumx=sum(xs)
sumx2=sum(xi**2 for xi in xs )
sumy=sum(logy)
sum2y=sum(yi**2 for yi in logy )
sumxy=sum(xi*yi for xi,yi in zip(xs,logy))
a1=(n*sumxy-sumx*sumy)/(n*sumx2-sumx**2)
a2=sumy/n-a1*sumx/n
line=[a1*delta*i+a2 for i in range(n0)]
xaxis=[delta*i for i in range(n0)]
plt.plot(xaxis,line,color='green')
plt.scatter(xs,logy,color='red')
```
