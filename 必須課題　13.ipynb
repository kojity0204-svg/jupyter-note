```python
import random 
import matplotlib.pyplot as plt
import math
fig=plt.figure(figsize=(16,10))
fig.suptitle('ramdom_walk')
step=100
random.seed(10)
c=[]
fig=plt.gcf()
plt.subplot(2,2,1)
plt.xlabel('step')
plt.ylabel('value')
for t in range(100):
    a=[0]
    d=[]
    b=0
    for s in range(step):
        if random.random()<0.5:
            b+=1
        else: 
            b-=1
        a.append(b)
        d.append(b)
    c.append(d)
    plt.plot(a)
fig=plt.gcf()
average=[]
for i in range(100):
    s=0
    for t in range(100):
        s+=c[t][i]
    average.append(s/100)

print(average)
plt.subplot(2,2,2)
plt.ylabel('average')
plt.xlabel('step')
plt.plot(average)
plt.subplot(2,2,3)
plt.xlabel('step')
mean_square=[]
for i in range(step):
    s=0
    for t in range(100):
        s+=(c[t][i]-average[i])**2
    mean_square.append(s/100)
plt.plot(mean_square)
plt.ylabel('mean_square')
```

```python
random.seed(200)
fig=plt.figure(figsize=(16,10))
N=100
b=N
delta_t=0.01
c=10000
l=0.001
plt.subplot(2,2,1)
a=[]
for i in range(c):
    b=b-l*b*delta_t
    a.append(b)
plt.xlabel('time')
plt.ylabel('analytical value')
plt.plot(a)
plt.subplot(2,2,2)
plt.ylabel('simulation')
b=N
d=[]
for t in range (10000):
    c=b
    for i in range(c):
        if random.random()<l:
            b-=1
    d.append(b)
plt.xlabel('time')
plt.ylabel('simulation')
plt.plot(d)
```

```python
N=1000000
random.seed(20)
xlavel=0
ylavel=0
x_outside=[]
y_outside=[]
x_inside=[]
y_inside=[]
a=0
for i in range(N):
    xlavel=random.random()
    ylavel=random.random()
    if ylavel<math.sqrt(1-xlavel**2):
        a+=1
        x_outside.append(xlavel)
        y_outside.append(ylavel)
    else:
        x_inside.append(xlavel)
        y_inside.append(ylavel)
plt.scatter(x_inside,y_inside,color='red')
plt.scatter(x_outside,y_outside,color='green')
print(4*a/N)

```
