```python
import math
import matplotlib.pyplot as plt
a=([0,0],[0,0])
b=([0,0],[0,0])
c=0
e=0.0001
def determinant(a):
    if abs(a[0][0]*a[1][1]-a[0][1]*a[1][0])<e:
        return "aは正則関数でありません"
    else:
        c=a[0][0]*a[1][1]-a[0][1]*a[1][0]
        b[0][0]=a[1][1]/c
        b[1][1]=a[0][0]/c
        b[1][0]=-a[1][0]/c
        b[0][1]=-a[0][1]/c
        return b
    

print(determinant(([1,2],[2,1])))
k=0.1
list_x=[x/10. for x in range(-25,25)]
list_y1=list(map(lambda x:math.cos(x)*math.cos(x),list_x))
list_y2=list(map(lambda x:0.5*k*x**2,list_x))
plt.plot(list_x,list_y1,color='green',label='cos(x)**2')
plt.plot(list_x,list_y2,color='red',label='kx**2/2')
plt.title('function')#グラフタイトル
plt.xlabel('x axis')#x軸
plt.ylabel('y axis')#y軸
plt.legend()#凡例
plt.show()
```
