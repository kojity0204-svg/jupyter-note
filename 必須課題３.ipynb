```python
i=0
list_f=[1,1]
def fibonacci (n):
    if n==0 or n==1:
        list_f[1]=1
    else:
        list_f.extend([0]*(n-1))
        for i in range(n-1):
            list_f[i+2]=list_f[i]+list_f[i+1]
    return list_f[n]
    
print(fibonacci (10000))
f=0
j=0
i=0
k=0
def innerproduct(a,b):
    global f
    if len(a)==len(b):
        for j in range(len(a)):
            f+=a[j]*b[j]
    else:
        print("error")
    return f
print(innerproduct([1,3],[2,3]))           
R=[0]*3
def exproduct(a,b):
    for i in range(3):
       for j in range(3):
         for k in range(3):
             global R
             if j!=i and k!=i and j>k:
                R[i]=a[k]*b[j]-b[k]*a[j]
       if i==1:
           R[i]=-R[i]
    return R 

print(exproduct([1,0,0],[0,1,0]))

```
