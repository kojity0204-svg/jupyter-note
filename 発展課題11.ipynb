```python
import random
import matplotlib.pyplot as plt
from matplotlib import animation,rc
from IPython.display import HTML
rc('animation', html='jshtml')
fig,ax=plt.subplots()
m,n=32,40
v=[]
for i in range(m):
    x_row=[]
    for j in range(n):
        x_row.append(random.randint(1,20))
    v.append(x_row)
m,n=32,40
xs=[]
file=open('file3.txt','r')
text=file.read()
file.close()
for line in text.split():
    x_row=[]
    for value in line.split(','):
        x_row.append(int(value))
    xs.append(x_row)
def surrounding(i,j):
    count=[]
    for p in range(max(0,i-1),min(m,i+2)):
        for q in range(max(0,j-1),min(n,j+2)):
            if p!=i or q!=j:
                count.append(xs[p][q])
    return count
def drawing():
    image=plt.pcolor(xs)
    return [image]
def draw():
    ax.invert_yaxis()
    plt.pcolor(xs)
    plt.show()
#draw()
history=[drawing()]
step=20
ts=list(range(step+1))
def next_state(current,surrounding,a):                  #vはあらかじめそれぞれのcellにランダムで割り当ててそこから動かす
    if current==3:
        return 4
    elif current==4:
        return 4
    elif current==0:
        return 0
    elif current==2:
        if a>=10: 
            return 3
        else: return 2
    elif current==1:
        if 3 in surrounding:
            if a>=10:
                return 3
            else: return 2
        else: return 1
for t in range(step):
    xs_next=[[0]*n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            xs_next[i][j]=next_state(xs[i][j],surrounding(i,j),v[i][j])
            if xs_next[i][j]==2:
                v[i][j]+=1
    xs=xs_next
    print(xs)
    history.append(drawing())
print(history)
#draw()
anim=animation.ArtistAnimation(fig,history,interval=500,blit=True)
plt.close()
anim

    
    
```
