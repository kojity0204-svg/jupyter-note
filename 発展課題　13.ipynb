```python
import random
from matplotlib import animation,rc
import matplotlib.pyplot as plt
import math
from IPython.display import HTML
rc('animation', html='jshtml')
fig,ax=plt.subplots()
step=100
random.seed(10)
c=[0]
b=[0]
x=0
y=0
def drawing():
    image=plt.scatter(x,y)
    return[image]
fig=plt.gcf()
history=[]
for i in range(step):
    x=c[i]
    y=b[i]
    if random.random()>0.5:
        x+=0.5
        c.append(x)
    else:
        x-=0.5
        c.append(x)
    if random.random()>0.5:
        y+=0.5
        b.append(y)
    else:
        y-=0.5
        b.append(y)
    history.append(drawing())
anim=animation.ArtistAnimation(fig,history,interval=100,blit=True)
plt.close()
anim
    
```

```python
N=100 
import random              #Nは粒子数
random.seed(10)
c=[]
b=[]
x=0
y=0
for n in range(N):
    c1=[0]
    b1=[0]                           #c1,b1はそれぞれの粒子の運動状態を仮止めするための数列
    for i in range(step):
        x=c1[i]
        y=b1[i]
        if random.random()<0.5:
            x+=0.5
            c1.append(x)
        else:
            x-=0.5
            c1.append(x)
        if random.random()<0.5:
            y+=0.5
            b1.append(y)
        else:
            y-=0.5
            b1.append(y)
    c.append(c1)                             #c,bがそれぞれx,yに対する値
    b.append(b1)
averagex=[]
for i in range(step):
    a=0
    for t in range(N):
        a+=c[t][i]
    averagex.append(a/N)
averagey=[]
for i in range(step):
    a=0
    for t in range(N):
            a+=b[t][i]
    averagey.append(a/N)
meanx=[]
for i in range(step):
    a=0
    for t in range(N):
        a+=c[t][i]-averagex[i]
    meanx.append(a/N)
meany=[]
for i in range(step):
    a=0
    for t in range(N):
        a+=b[t][i]-averagey[i]
    meany.append(a/N)

meansquare=[]
for i in range(step):
    a=0
    for t in range(N):
        a+=(c[t][i]-averagex[i])**2+(b[t][i]-averagey[i])**2
    meansquare.append(a/N)
plt.plot(meanx,color='blue',label='meanx')
plt.plot(meany,color='red',label='meany')
plt.plot(meansquare,color='green',label='meansquare')
plt.legend()
plt.show()
```
