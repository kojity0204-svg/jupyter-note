```python
import math
import matplotlib.pyplot as plt
a=([0,0],[0,0])
b=([0,0],[0,0])
c=([0,0],[0,0])
def matrixproduct(a,b):
    c[0][0]=a[0][0]*b[0][0]+a[0][1]*b[1][0]
    c[1][0]=a[1][0]*b[0][0]+a[1][1]*b[1][0]
    c[0][1]=a[0][0]*b[0][1]+a[0][1]*b[1][1]
    c[1][1]=a[1][0]*b[0][1]+a[1][1]*b[1][1]
    return c
print(matrixproduct(([1,0],[0,1]),([1,2],[2,1])))
b=25
d=25
l=5
a_1=0
b_1=0
a_2=0
b_2=0
def L1(x):
    a_1=(b*math.pi)/l*math.sin(x)
    b_1=(math.sin(a_1)/a_1)**2
    return b_1

def L2(x):
    a_2=(d*math.pi)/l*math.sin(x)
    b_2=(math.cos(a_2))**2
    return b_2

def L3(x):
    return L1(x)*L2(x)

list_x1=[-math.pi/4+math.pi*x/6400. for x in range(1600)]
list_x2=[math.pi*(x+1)/6400. for x in range (1600)]
list_y1=list(map(lambda x:L1(x),list_x1))
list_y2=list(map(lambda x:L2(x),list_x1))
list_y12=list(map(lambda x:L1(x),list_x2))
list_y22=list(map(lambda x:L2(x),list_x2))
list_y31=list(map(lambda x: L3(x),list_x1))
list_y32=list(map(lambda x: L3(x),list_x2))
plt.plot(list_x1,list_y1,color='green',label='I1')
plt.plot(list_x1,list_y2,color='red',label='I2')
plt.plot(list_x2,list_y12,color='green',label='I1')
plt.plot(list_x2,list_y22,color='red',label='I2')
plt.plot(list_x1,list_y31,color='blue',label='I3')
plt.plot(list_x2,list_y32,color='blue',label='I3')
plt.title('function')#グラフタイトル
plt.xlabel('x axis')#x軸
plt.ylabel('y axis')#y軸
plt.legend()#凡例
plt.show()
```
