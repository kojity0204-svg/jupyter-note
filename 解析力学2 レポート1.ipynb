```python
import math 
import matplotlib.pyplot as plt
n=10000
delta=0.01
x0=0
p0=0.1
m=1
w=1
def dhpdp(p):
    return p/m
def dhq1dq(q):
    return m*(w**2)*q
def dhq2dq(q,p):
    return m*(w**2)*q+m*(w**2)*delta*dhpdp(p)
xs1=[]
ps1=[]
p=p0
x=x0
xs1.append(x0)
ps1.append(p0)
for i in range (n):
    k1=dhpdp(p)
    k2=dhq1dq(x)
    p=p-delta*k2
    x=x+delta*k1
    xs1.append(x)
    ps1.append(p)
plt.subplot(2,2,1)
plt.plot(xs1,ps1)
xs2=[]
ps2=[]
p=p0
x=x0
xs2.append(x0)
ps2.append(p0)
for i in range (n):
    k1=dhpdp(p)
    k2=dhq2dq(x,p)
    p=p-delta*k2
    x=x+delta*k1
    xs2.append(x)
    ps2.append(p)
plt.subplot(2,2,2)
plt.plot(xs2,ps2)
```
